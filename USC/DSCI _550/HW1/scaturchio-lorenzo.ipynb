```python
# Importing Packages
import numpy as np
import pandas as pd
import seaborn as sns

# Loading in the data set
df = pd.read_csv('flights.csv')

# Eliminating any values that are negative and turning them to 0
df.loc[df["dep_delay"] < 0, "dep_delay"] = 0
df.loc[df["arr_delay"] < 0, "arr_delay"] = 0

df

```

# 1. What is the average departure delay time per month and which month did we have the most departure delays overall?

```python
grouped_dep_delay = df[["month", "dep_delay"]].groupby("month").mean()
grouped_dep_delay
```

```python
grouped_dep_delay.sort_values('dep_delay', ascending=False).head(1)
```

# 2. What is the average arrival and departure delay per air carrier?


```python
grouped_air_carrier = df[["carrier", "dep_delay", "arr_delay"]].groupby("carrier").mean()
grouped_air_carrier
```

# 3. What were the top 5 flights with the worst departure delays?

```python
worst_flights = df[['flight', 'dep_delay']].sort_values('dep_delay', ascending=False)
worst_flights.head(5)
```

# 4. How much delay do I need to consider when I use AA in July? (based on average)

```python
AA_delay = df[['month', 'dep_delay', 'arr_delay']].loc[df["carrier"] == "AA"].loc[df["month"] == 7]
AA_delay.groupby('month').mean()
```
